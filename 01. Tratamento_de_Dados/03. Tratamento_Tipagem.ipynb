## Alterando a tipagem dos dados...

```python
df['Salario_Anual'] = df['Salario_Anual'].astype(float)
```

O erro acima evidencia que alguns registros estão como string por causa dos pontos que estão sendo utilizados de separadores

```python
# REMOVENDO OS . SEPARADORES
df['Salario_Anual'] = df['Salario_Anual'].str.replace('.', '', regex = False) # REGEX = FALSE garante que apenas o valor exato (".") seja substituído, quando é TRUE, o regex se comporta de acordo com seus caracteres especiais, nesse caso "." é um coringa e iria substituir todos os caracteres presentes nos registros com "."

# REFAZENDO A CONVERSÃO
df['Salario_Anual'] = df['Salario_Anual'].astype(float)
```

```python
df['Churn'] = df['Churn'].astype(int)
```

O erro acima indica que o pandas não pode converter os registros com erros / nulos (NA)
