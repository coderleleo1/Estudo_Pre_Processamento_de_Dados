```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Exemplo 1 ---> Números incorretos

```python
dados = {
    'Idade': [25, 30, 35, 400, 45, 50, 554, 60, 60, 24, 233, 70, 26]
}

df = pd.DataFrame(dados) # ---> CRIAÇÃO DO DATAFRAME
print(df)
```

O ideal é sempre primeiro verificar a integridade dos dados antes de apresentar a análise, ver se os dados são coesos com a realidade...

```python
media = df['Idade'].mean()
print(media)
```

## Exemplo 2 ---> Erros de Digitação

```python
dados = {
    'Idade': [25, 30, 35, 400, 45, 50, 554, 60, 60, 24, 233, 70, 26],
    'Cidade': ['São Paulo   ', 'Salvador', 'Curitiba', 'Sao Paulo', 'Salvador', 'Curitiba', 'Sao Paulo', 'Salvador', 'salvador', 'Curtiba','São Paulo','São  Paulo','salvador']
}

df = pd.DataFrame(dados)
print(df)
```

É importante verificar a digitação dos categóricos dos dados, ver acentuação, letras maiúsculas e minúsculas, erros de digitação e também espaços


```python
df.groupby('Cidade')['Idade'].mean().reset_index().sort_index(ascending = False)
```

Manter um padrão para dados de texto, preferencialmente não usar acentuação e também sempre procurar por erros de digitação

## Exemplo 3 ---> Dados incompletos e tipagem incorreta

```python
dados = {
    'Idade': [25, 30, 35, 400, 45, 50, 554, 60, 60, 24, 233, 70, 26],
    'Cidade': ['São Paulo', 'Salvador', 'Curitiba', 'Sao Paulo', 'Salvador', 'Curitiba', 'Sao Paulo', 'Salvador', 'salvador', 'Curtiba','São Paulo','São Paulo','salvador'],
    'Salario': ['1200,0', '0', '4000,0', '0', '5000,0', '1500,0', '0', '0', '0', '3000,0','2500,0','3000,0','0']
}

df = pd.DataFrame(dados)
print(df)
```

Nem sempre é possível de ter os dados 100% completos, seja por conta de erros ou pelo cliente não fornecer a informação, por isso é importante analisar campos nulos para não enviesar a análise.

Outro erro comum é registros com a tipagem incorreta, nesse caso, o salário está como uma string e não como float ou int, portanto, não é possível executar funções matemáticas nesse coluna

```python
# ERRO DE TIPAGEM
df.groupby('Cidade')['Salario'].mean().reset_index().sort_values(ascending = False)
```
