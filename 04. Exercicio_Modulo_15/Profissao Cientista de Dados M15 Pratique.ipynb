## Exercício C

### Coluna Gênero

```python
contagem_genero = df['Genero'].value_counts().reset_index(name = 'Contagem')
contagem_genero['Percentual'] = ((contagem_genero['Contagem'] / len(df)) * 100).round(2)

print('Valores Brutos: \n', contagem_genero, '\n')
fig = px.bar(
    contagem_genero,
    x = 'Genero',
    y = 'Percentual',
    title = 'Gêneros de Clientes [%]',
    template = 'plotly',
    width = 1000,
    height = 600,
    labels = {'Genero': 'Gênero'},
    color = 'Genero',
    color_discrete_map = {
        'Female': '#FD8AFF',
        'Male': '#6978FF'
    },
    text_auto = True
)

fig.update_layout(
    font = dict(
        size = 16
    ),
    hoverlabel = dict(
        font = dict(
            size = 14
        )
    )
)

fig.update_traces(
    textfont_size = 18,
    textposition = 'outside',
)

fig.show()
```

### **Insight**

Podemos observar que os gêneros dos clientes é bem balanceado, sendo **51% homens e 49% mulheres**

### Coluna Idoso

```python
# CRIANDO UM DATAFRAME PARALELO ---> MELHORAR A VISUALIZAÇÃO
df_idoso_tratado = df.copy()
df_idoso_tratado['Idoso'] = df_idoso_tratado['Idoso'].replace({0: 'Não Idoso', 1: 'Idoso'})

# CALCULANDO OS VALORES
contagem_idosos = df_idoso_tratado['Idoso'].value_counts().reset_index(name = 'Contagem')
contagem_idosos['Percentual'] = ((contagem_idosos['Contagem'] / len(df)) * 100).round(2)

print(f'Valores Brutos: \n {contagem_idosos}\n')

# PLOTANDO O GRÁFICO
fig = px.pie(
    contagem_idosos,
    values = 'Percentual',
    names = 'Idoso',
    template = 'plotly',
    hole = .5,
    title = 'Clientes Idosos [%]',
    width = 1000,
    height = 600,
    category_orders = {
        'Idoso': ['Idoso', 'Não Idoso']
    },
    color = 'Idoso',
    color_discrete_map = {
       'Idoso': '#6978FF',
       'Não Idoso': '#7f7f7f'
    }
)

fig.update_layout(
    font = dict(
        size = 20
    ),
    legend = dict( # ALTERANDO A LEGENDA DO GRÁFICO
    x = 0.8, # POSIÇÃO HORIZONTAL
    y = 1, # POSIÇÃO VERTICAL
    ),
    hoverlabel = dict(
        font = dict(
            color = 'white',
            size = 16
        )
    )
)

fig.update_traces(
    textfont = dict(
        color = 'white',
    )
)

fig.show()
```

### **Insight**

Podemos observar que 83% do clientes não são pessoas idosas

### Coluna Tempo_Como_Cliente

```python
df_tempo = df.copy()

valor_maximo = df_tempo['Tempo_como_Cliente'].max()
bins = list(range(0, valor_maximo + 7, 6))
labels = [f'{b} a {b + 5}' for b in bins[:-1]]

df_tempo['Faixa'] = pd.cut(
    df_tempo['Tempo_como_Cliente'],
    bins = bins,
    labels = labels,
    right = False,  # <--- CORREÇÃO: Mudar de True para False
    include_lowest = True
)

tabela_faixas = df_tempo['Faixa'].value_counts().sort_index().reset_index(name = 'Contagem')
tabela_faixas['Percentual'] = ((tabela_faixas['Contagem'] / len(df_tempo)) * 100).round(2)
print(tabela_faixas, '\n')

fig = px.bar(
    tabela_faixas,
    x = 'Faixa',
    y = 'Percentual',
    template = 'plotly',
    title = 'Distribuição de Clientes [%] Por Faixa de Tempo',
    width = 1000,
    height = 600
)

fig.update_layout(
    bargap = 0.05,
    font = dict(
        size = 14
    )
)

fig.show()
```

## Exercício A

### Relação Tipo de Contrato Com Churn

```python
#Seu código aqui
df_agrupado_contrato_churn = df.groupby(['Churn', 'Tipo_Contrato']).size().reset_index(name = 'Contagem')
qtd_clientes_churn = df_agrupado_contrato_churn.groupby('Churn')['Contagem'].transform('sum')

df_agrupado_contrato_churn['Percentual'] = (df_agrupado_contrato_churn['Contagem'] / qtd_clientes_churn * 100).round(2)


print(df_agrupado_contrato_churn, '\n')

fig = px.bar(
    df_agrupado_contrato_churn,
    x = 'Churn',
    y = 'Percentual',
    template = 'plotly',
    width = 1000,
    height = 600,
    color = 'Tipo_Contrato',
    title = 'Relação Churn e Tipo de Contrato'
)

fig.update_layout(
    hoverlabel = dict(
       font = dict(
            color = 'white',
            size = 16
        )
    ),
    font = dict(
        size = 18
    )
)

fig.show()
```

### **Insight**
Podemos observar que clientes com contrato mês a mês possuem forte propensão a serem churns. Cerca de 89% da nossa base Churn são clientes com contrato de mensal

## Relação Churn e Método de Pagamento

```python
df_agrupado_churn_metodo_pgto = df.groupby(['Churn', 'Metodo_Pagamento']).size().reset_index(name = 'Contagem')
qtd_clientes_churn = df_agrupado_churn_metodo_pgto.groupby('Churn')['Contagem'].transform('sum')

df_agrupado_churn_metodo_pgto['Percentual'] = (df_agrupado_churn_metodo_pgto['Contagem'] / qtd_clientes_churn * 100).round(2)
print(df_agrupado_churn_metodo_pgto, '\n')

fig = px.bar(
    df_agrupado_churn_metodo_pgto,
    x = 'Churn',
    y = 'Percentual',
    template = 'plotly',
    color = 'Metodo_Pagamento',
    width = 1000,
    height = 600,
    title = 'Relação Churn e Método de Pagamento'
)

fig.show()
```

### **Insight**
Cerca de 58% da base de clientes churn possuem como método de pagamento o cheque eletrônico

## Relação do Tempo Como Cliente e Churn

```python
df_tempo = df.copy() # Criando uma cópia do DataFrame

# DEFININDO AS FAIXAS TEMPORAIS
valor_maximo = df_tempo['Tempo_como_Cliente'].max()
bins = list(range(0, valor_maximo + 7, 6))     # Intervalos
labels = [f'{b} a {b + 5}' for b in bins[:-1]] # Escrita das faixas

df_tempo['Faixas'] = pd.cut(
    df_tempo['Tempo_como_Cliente'],
    bins = bins,
    labels = labels,
    right = False,
    include_lowest = True
)

# CALCULANDO A QUANTIDADE DE CHURN POR FAIXA E TIRANDO O PERCENTUAL DE REPRESENTATIVIDADE
df_agrupado_tempo_churn = df_tempo.groupby(['Churn', 'Faixas']).size().reset_index(name = 'Contagem')
clientes_churn = df_agrupado_tempo_churn.groupby('Faixas')['Contagem'].transform('sum')
df_agrupado_tempo_churn['Percentual'] = (df_agrupado_tempo_churn['Contagem'] / clientes_churn * 100).round(2)

# PLOTANDO O GRÁFICO
fig = px.bar(
    df_agrupado_tempo_churn,
    x = 'Faixas',
    y = 'Percentual',
    color = 'Churn',
    color_discrete_map = {
        'Yes': '#FF5C5C',
        'No': '#7f7f7f'
    },
    template = 'plotly',
    width = 1000,
    height = 600,
    category_orders = {'Churn': ['Yes', 'No']},
    title = 'Percentual de Churn por Faixa de Tempo',
)

fig.show()
```

### **Insight**
Conforme mais tempo os clientes passam no nosso negócio, menor é a propensão de serem Churns

## Relação Valor Pagamento e Churn

```python
df_pgto = df.copy()

# DEFININDO AS FAIXAS DE PAGAMENTO
valor_maximo = df_pgto['Pagamento_Mensal'].max()
bins = list(range(0, int(valor_maximo) + 10,11))
labels = [f'{b} a {b + 10}' for b in bins[:-1]]

df_pgto['Faixas'] = pd.cut(
    df_pgto['Pagamento_Mensal'],
    bins = bins,
    labels = labels,
    include_lowest = True,
    right = False,
)

# CALCULANDO CLIENTES CHURN POR FAIXA DE VALOR
df_agrupado = df_pgto.groupby(['Churn', 'Faixas']).size().reset_index(name = 'Contagem')
clientes = df_agrupado.groupby('Faixas')['Contagem'].transform('sum')
df_agrupado['Percentual'] = (df_agrupado['Contagem'] / clientes * 100).round(2)

fig = px.bar(
    df_agrupado,
    x = 'Faixas',
    y = 'Percentual',
    color = 'Churn',
    template = 'plotly',
    category_orders = {
        'Churn': ['Yes', 'No'],
    },
    color_discrete_map = {
        'Yes': '#FF5C5C',
        'No': '#7f7f7f'
    },
    width = 1000,
    height = 600,
)

fig.show()
```

### **Insight**
Clientes com faixas de mensalidade mais altas tendem a serem Churns
