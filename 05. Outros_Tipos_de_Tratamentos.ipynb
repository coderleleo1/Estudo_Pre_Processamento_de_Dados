```python
import pandas as pd
```

```python
df = pd.read_csv('Dados/CHURN_CREDIT_MOD14_TRATADO.csv')
```

### Padronizando ---> Renomeando a coluna Balance para Balanco

```python
df.rename(columns = {'Balance': 'Balanco'}, inplace = True)
df.head(5)
```

### Tratando os valores da coluna Pais

```python
df['Pais'].value_counts()
```

```python
# TRANSFORMANDO TODOS OS DADOS DA COLUNA EM CAIXA BAIXO (MINÚSCULA)
df['Pais'] = df['Pais'].str.lower()
df['Pais'].unique()
```

```python
# TRANSFORMANDO TODOS OS DADOS DA COLUNA EM CAIXA BAIXO (MAIÚSCULA)
df['Pais'] = df['Pais'].str.upper()
df['Pais'].unique()
```

Padronizando as categorias

```python
df['Pais'] = df['Pais'].replace('GERMANI', 'GERMANY')
df['Pais'].unique()
```

### Verificando a coluna de gênero

```python
df['Genero'].unique()
```

### Retirando as colunas de Sobrenome e CustomerId

```python
df.drop(columns = ['Sobrenome'], inplace = True)
df.drop(columns = ['CustomerId'], inplace = True)
df.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
df.head(5)
```

### Salvando o DataFrame

```python
df.to_csv('Dados/CHURN_CREDIT_MOD14_TRATADO.csv', index = False, encoding = 'utf-8')
df_tratado = pd.read_csv('Dados/CHURN_CREDIT_MOD14_TRATADO.csv')
df_tratado.head(5)
```
